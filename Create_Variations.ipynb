```python
import os
import json
import requests
import openai
from base64 import b64decode
from io import BytesIO
from PIL import Image
```

```python
openai.api_key = os.environ.get("OPENAI_API_KEY")
```

#### Parameters

- Reference image -  PNG format with less than 4MB, and square.


- n - 1 to 10


- size - 256x256, 512x512, or 1024x1024


- response_format - url or b64_json

```python
size = "512x512"
```

```python
ref_img_path = "images/image_url_0.png"

response = openai.Image.create_variation(
    image=open(ref_img_path, mode="rb"), n=2, size=size, response_format="b64_json",
)
print(response)
```

```python
for index, image_dict in enumerate(response["data"]):
    img = b64decode(image_dict["b64_json"])
    image_file = "images/image_var_" + str(index) + ".png"
    with open(image_file, mode="wb") as img_file:
        img_file.write(img)
```
